```python
data_for_tree = pd.get_dummies(data)
data_for_tree.head()
#ужаснемся названиям столбцов
```

```python
#переназовем некоторые громоздкие колонки
data_for_tree = data_for_tree.rename(columns = {'cp_atypical angina' : 'cp_atypical_angina',
                              'cp_non-anginal pain' : 'cp_non-anginal_pain',
                              'cp_typical angina' : 'cp_typical_angina', 
                              'restecg_having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV' : 'restecg_ST-T', 
                              'restecg_showing probable or definite left ventricular hypertrophy by Estes criteria' : 'restecg_ventricular'})
```

```python
#проверим, что всё правильно сделали
data_for_tree.head()
```

```python
X = data_for_tree.drop('num', axis=1)
y = data_for_tree.num
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)
```

```python
clf.score(X, y)
```

```python
from IPython.display import SVG
from graphviz import Source
from IPython.display import display
```

```python
from IPython.display import HTML
style = "<style>svg{width:10% !important;height:10% !important;}</style>"
HTML(style)
graph = Source(tree.export_graphviz(clf, out_file=None
   , feature_names=list(X), class_names=['Diseased', 'Healthy'] 
   , filled = True))

display(SVG(graph.pipe(format='svg')))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```
